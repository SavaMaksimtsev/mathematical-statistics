```python
# Задача: найти коэффициент корреляции между расстоянием и временем полета; построить точечную диаграмму в осях distance (х) и air_time (y); найти коэффициенты линейной регрессии и нанести полученную прямую на график; интерпретировать смысл коэффициентов полученной модели
# Библиотеки все те же
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

df = pd.read_csv('flights_NY.csv').dropna(axis=0)

# Вычислил коэффициент корреляции между расстоянием и временем полёта
corr_coef, _ = pearsonr(df['distance'].values, df['air_time'].values)
print(f'Коэффициент корреляции: {corr_coef:.2f}')

# Построил точечную диаграмму
plt.scatter(df['distance'], df['air_time'], color='#6a9f72')
plt.xlabel('Расстояние (в милях)')
plt.ylabel('Время полёта в (минутах)')
plt.title('Корреляция между расстоянием и временем полёта')

# Посчитал коэффициенты линейной регрессии
X = df[['distance']].values
y = df['air_time'].values
coefficients = np.polyfit(X.ravel(), y, 1)
polynomial = np.poly1d(coefficients)

# Наложил прямую на диаграмму
xp = np.linspace(X.min(), X.max(), 100)
plt.plot(xp, polynomial(xp), "-", color="#a72217")

plt.show()

# Полученные коэффициенты позволяют сказать, что с эфирное время увеличивается на 0,1 минуты
print(f'Уклон: {coefficients[0]:.2f} минут на милю')
```
